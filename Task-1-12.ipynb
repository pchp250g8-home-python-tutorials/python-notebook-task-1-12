При каких натуральных значениях n выражение (3 * n ** 2 - 16 * n + 21)  /  (n - 3) является натуральным числом?

```python
from sympy.abc import n
from sympy import apart
from sympy import pdiv
from sympy import solve

print(pdiv(3 * n ** 2 - 16 * n + 21, n - 3))
print(apart((3 * n ** 2 - 16 * n + 21) / (n - 3)))
print(solve(n - 3 > 0, n))

```
